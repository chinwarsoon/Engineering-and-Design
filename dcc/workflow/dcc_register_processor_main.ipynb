```python
##P.4 Load schema and parameters with correct precedence
# PRECEDENCE: JSON file parameters (highest) → CLI arguments (medium) → Native defaults (fallback)

status_print("Loading schema and parameters with correct precedence...")

try:
    # STEP 1: Load JSON file parameters (HIGHEST PRIORITY)
    json_parameters = {}
    schema_register_file = cli_args.get('schema_register_file') or native_defaults['schema_register_file']
    
    if schema_register_file and Path(schema_register_file).exists():
        # Use CLI-provided or native default schema register
        with open(schema_register_file, 'r', encoding='utf-8') as f:
            schema_register = json.load(f)
        json_parameters = schema_register.get('parameters', {})
        status_print(f"✓ JSON config loaded from: {schema_register_file}")
    else:
        # No schema register file found - use empty JSON parameters
        json_parameters = {}
        debug_print(f"Schema register file not found: {schema_register_file}, using empty JSON parameters")
    
    # STEP 2: Apply CLI arguments (MEDIUM PRIORITY) - overrides JSON
    # Start with JSON parameters (highest priority)
    effective_parameters = json_parameters.copy()
    
    # CLI arguments override JSON parameters
    if cli_args:
        status_print("Applying CLI arguments (medium priority):")
        for param, value in cli_args.items():
            effective_parameters[param] = value
            status_print(f"  ✓ {param} = {value} (overrides JSON if present)")
    
    # STEP 3: Apply native defaults (LOWEST PRIORITY - final fallback)
    # Only apply defaults for parameters not already set
    for param, value in native_defaults.items():
        if param not in effective_parameters:
            effective_parameters[param] = value
            debug_print(f"  ✓ {param} = {value} (native default)")
    
    # Update global variables with effective parameters
    for param, value in effective_parameters.items():
        globals()[param] = value
    
    # Load schema path from master registry (using effective registry_path)
    with open(effective_parameters['registry_path'], 'r', encoding='utf-8') as f:
        registry = json.load(f)
    
    schema_file = registry['document_types']['DCC_SUBMITTAL_ENHANCED']['schema_file']
    schema_path = Path.cwd().parent / schema_file
    status_print(f"✓ Schema path from registry: {schema_path}")
    
    status_print("Parameter precedence applied successfully:")
    status_print("  1. JSON file parameters (highest priority)")
    status_print("  2. CLI arguments (medium priority)")  
    status_print("  3. Native defaults (final fallback)")
    
    # Show precedence summary
    if json_parameters:
        status_print(f"  JSON parameters loaded: {len(json_parameters)} items")
    if cli_args:
        status_print(f"  CLI arguments applied: {len(cli_args)} items")
    status_print(f"  Final effective parameters: {len(effective_parameters)} items")
    
    # Check debug mode AFTER precedence is applied
    if effective_parameters.get('debug_dev_mode', False):
        status_print(f"Debug/Dev Mode: {effective_parameters.get('debug_dev_mode')}")
        status_print(f"Environment: {'Google Colab' if is_colab else 'Local'}")
        status_print(f"Upload File: {effective_parameters.get('upload_file_name')}")
        status_print(f"Download Path: {effective_parameters.get('download_file_path')}")
        status_print(f"Registry Path: {effective_parameters.get('registry_path')}")
        status_print(f"Schema Register File: {effective_parameters.get('schema_register_file')}")
        
        
    
    cell_done("Schema and parameters loaded with correct precedence")
    
except Exception as e:
    status_print(f"❌ Error loading schema: {e}")
    raise

debug_print("All parameters loaded and listed below:")
for key, value in effective_parameters.items():
    debug_print(f"{key}: {value}")

print("=" * 80)
```

```python
print("\n📂 Step 2: Load Excel Data & Column Mapping")
print("=" * 50)

try:
    # Force reload the universal_column_mapper module to get the latest fixes
    import importlib
    import universal_column_mapper
    importlib.reload(universal_column_mapper)
    from universal_column_mapper import UniversalColumnMapper
    
    # Recreate mapper with updated module
    mapper = UniversalColumnMapper()
    mapper.load_main_schema(schema_path)
    
    # Use parameters with proper precedence (already applied in Step 1)
    file_to_load = effective_parameters.get('upload_file_name')
    sheet_to_use = effective_parameters.get('upload_sheet_name')
    header_row = effective_parameters.get('header_row_index')
    start_col = effective_parameters.get('start_col', 'A')
    end_col = effective_parameters.get('end_col', None)
    
    print(f"📁 Loading Excel file: {file_to_load}")
    print(f"📋 Sheet: {sheet_to_use or 'First sheet'}")
    print(f"📊 Header row: {header_row}")
    print(f"📏 Column range: {start_col} → {end_col or 'All columns'}")
    
    if not Path(file_to_load).exists():
        raise FileNotFoundError(f"Excel file not found: {file_to_load}")
    
    # Convert column letters to numbers
    def col_letter_to_num(col_letter):
        if isinstance(col_letter, int):
            return col_letter
        col_letter = str(col_letter).upper()
        result = 0
        for char in col_letter:
            if char.isalpha():
                result = result * 26 + (ord(char) - ord('A') + 1)
        return result
    
    def col_num_to_letter(col_num):
        result = ""
        while col_num > 0:
            col_num -= 1
            result = chr(col_num % 26 + ord('A')) + result
            col_num //= 26
        return result
    
    # Calculate column range for pandas
    start_num = col_letter_to_num(start_col)
    if end_col:
        end_num = col_letter_to_num(end_col)
        col_range = f"{col_num_to_letter(start_num)}:{col_num_to_letter(end_num)}"
        use_cols = list(range(start_num - 1, end_num))  # pandas is 0-indexed
        print(f"📏 Loading columns {col_range} ({len(use_cols)} columns)")
    else:
        use_cols = None  # Load all columns
        print(f"📏 Loading all columns from {col_num_to_letter(start_num)}")
    
    # Read the Excel file with column range
    if sheet_to_use:
        df_raw = pd.read_excel(
            file_to_load, 
            sheet_name=sheet_to_use, 
            header=header_row,
            usecols=use_cols
        )
        print(f"✅ Loaded sheet: {sheet_to_use}")
    else:
        df_raw = pd.read_excel(
            file_to_load, 
            header=header_row,
            usecols=use_cols
        )
        print(f"✅ Loaded Excel file (first sheet)")
    
    print(f"📊 Initial data shape: {df_raw.shape}")
    
    # Clean the data by dropping blank rows, blank columns, and columns without titles
    print(f"\n🧹 Cleaning Excel data...")
    
    # Store original shape for reporting
    original_shape = df_raw.shape
    original_columns = df_raw.columns.tolist()
    
    # 1. Drop columns without titles (unnamed/empty headers)
    cols_to_drop = []
    for col in df_raw.columns:
        if pd.isna(col) or str(col).strip() == '' or str(col).startswith('Unnamed:'):
            cols_to_drop.append(col)
    
    if cols_to_drop:
        df_raw = df_raw.drop(columns=cols_to_drop)
        print(f"  🗑️  Dropped {len(cols_to_drop)} columns without titles:")
        for col in cols_to_drop:
            print(f"     - {col}")
    else:
        print(f"  ✅ No columns without titles found")
    
    # 2. Drop completely blank columns
    blank_cols = []
    for col in df_raw.columns:
        if df_raw[col].isna().all():
            blank_cols.append(col)
    
    if blank_cols:
        df_raw = df_raw.drop(columns=blank_cols)
        print(f"  🗑️  Dropped {len(blank_cols)} completely blank columns:")
        for col in blank_cols:
            print(f"     - {col}")
    else:
        print(f"  ✅ No completely blank columns found")
    
    # 3. Drop completely blank rows
    blank_rows_before = len(df_raw)
    df_raw = df_raw.dropna(how='all')  # Drop rows where all values are NaN
    blank_rows_dropped = blank_rows_before - len(df_raw)
    
    if blank_rows_dropped > 0:
        print(f"  🗑️  Dropped {blank_rows_dropped} completely blank rows")
    else:
        print(f"  ✅ No completely blank rows found")
    
    # 4. Drop rows where all key columns are blank (optional - more aggressive cleaning)
    # This removes rows that might have some data but are essentially empty
    non_null_threshold = len(df_raw.columns) * 0.1  # At least 10% of columns should have data
    sparse_rows_before = len(df_raw)
    df_raw = df_raw.dropna(thresh=non_null_threshold)
    sparse_rows_dropped = sparse_rows_before - len(df_raw)
    
    if sparse_rows_dropped > 0:
        print(f"  🗑️  Dropped {sparse_rows_dropped} sparse rows (<10% data)")
    else:
        print(f"  ✅ No sparse rows found")
    
    print(f"📊 Cleaned data shape: {df_raw.shape}")
    print(f"📊 Data reduction: {original_shape[0]} → {df_raw.shape[0]} rows, {original_shape[1]} → {df_raw.shape[1]} columns")
    
    # Perform column mapping using universal column mapper
    print(f"\n🔄 Mapping columns to schema names...")
    
    # Create mapping result using the correct method name
    mapping_result = mapper.detect_columns(df_raw.columns)
    
    print(f"✅ Column mapping completed:")
    print(f"  Match rate: {mapping_result['match_rate']:.1%}")
    print(f"  Mapped columns: {mapping_result['matched_count']}")
    print(f"  Unmapped columns: {mapping_result['total_headers'] - mapping_result['matched_count']}")
    
    # Rename columns based on mapping
    df_renamed = mapper.rename_dataframe_columns(df_raw, mapping_result)
    
    def get_scalar_null_count(series_or_scalar):
        """Helper function to safely extract scalar null count from various pandas return types"""
        if isinstance(series_or_scalar, (int, float)):
            return series_or_scalar
        elif hasattr(series_or_scalar, 'iloc'):
            # pandas Series or Index
            if len(series_or_scalar) == 1:
                return series_or_scalar.iloc[0]
            else:
                return series_or_scalar.sum()  # Sum all values if multiple
        elif hasattr(series_or_scalar, '__len__') and len(series_or_scalar) == 1:
            return series_or_scalar[0]
        else:
            # Last resort - try to convert
            try:
                return int(series_or_scalar)
            except (TypeError, ValueError):
                return 0
    
    # Analyze duplicate mapped column names
    print(f"\n🔍 Duplicate Column Mapping Analysis:")
    print("=" * 80)
    
    # Create a mapping of raw columns to their mapped names
    raw_to_mapped = {}
    for i, raw_col in enumerate(df_raw.columns):
        if i < len(df_renamed.columns):
            mapped_col = df_renamed.columns[i]
            raw_to_mapped[raw_col] = mapped_col
    
    # Find duplicates by grouping by mapped column names
    mapped_to_raws = {}
    for raw_col, mapped_col in raw_to_mapped.items():
        if mapped_col not in mapped_to_raws:
            mapped_to_raws[mapped_col] = []
        mapped_to_raws[mapped_col].append(raw_col)
    
    # Show only columns that have duplicates
    duplicates_found = False
    for mapped_col, raw_cols in mapped_to_raws.items():
        if len(raw_cols) > 1:
            duplicates_found = True
            print(f"🔄 '{mapped_col}' is mapped from {len(raw_cols)} raw columns:")
            for raw_col in raw_cols:
                print(f"   • {raw_col}")
            print()
    
    if not duplicates_found:
        print("✅ No duplicate mapped column names found")
    
    print("=" * 80)
    
    # Print detailed column information with SIMPLE formatting to avoid corruption
    print(f"\n📋 Detailed Column Information:")
    print("=" * 120)
    print(f"{'#':>3} {'Raw Column':<30} {'Mapped Column':<30} {'Data Type':<12} {'Null Count':<10} {'Null %':<8}")
    print("=" * 120)
    
    # Show column comparison with data types and null counts
    for i, raw_col in enumerate(df_raw.columns):
        renamed_col = df_renamed.columns[i] if i < len(df_renamed.columns) else f"Column_{i}"
        
        # Get data type using proper pandas syntax
        dtype = str(df_renamed.dtypes[renamed_col])
        
        # Get null count using robust helper function
        null_series = df_renamed[renamed_col].isna().sum()
        null_count = get_scalar_null_count(null_series)
        null_pct = (null_count / len(df_renamed)) * 100
        
        # Get mapping status
        if raw_col in mapping_result['detected_columns']:
            mapped_name = mapping_result['detected_columns'][raw_col]['column_definition'].get('name', 'Unknown')
            if mapped_name != 'Unknown':
                mapping_suffix = "✅"
            else:
                mapping_suffix = "⚠️"
        else:
            mapping_suffix = "❌"
        
        # Format for display with truncation to avoid corruption
        raw_display = (raw_col[:28] + "..") if len(raw_col) > 28 else raw_col
        renamed_display = (renamed_col[:28] + "..") if len(renamed_col) > 28 else renamed_col
        dtype_display = (dtype[:10] + "..") if len(dtype) > 10 else dtype
        
        print(f"{i+1:>3} {raw_display:<30} {renamed_display:<30} {dtype_display:<12} {null_count:>8,} ({null_pct:>5.1f}%)")
    
    print("=" * 120)
    
    # Show mapping details for successfully mapped columns
    if mapping_result['detected_columns']:
        print(f"\n📋 Successful Column Mappings:")
        for original, mapping_info in mapping_result['detected_columns'].items():
            mapped_name = mapping_info['column_definition'].get('name', 'Unknown')
            if mapped_name != 'Unknown':
                print(f"  ✅ {original} → {mapped_name}")
            else:
                print(f"  ⚠️ {original} → Auto-detected")
    
    # Show unmapped columns
    if mapping_result['unmatched_headers']:
        print(f"\n⚠️ Unmapped Columns:")
        for col in mapping_result['unmatched_headers']:
            print(f"  - {col}")
    
    # Summary statistics
    print(f"\n📊 Data Quality Summary:")
    print(f"  Total rows: {len(df_renamed):,}")
    print(f"  Total columns: {len(df_renamed.columns)}")
    
    # Show columns with high null counts
    high_null_cols = []
    for col in df_renamed.columns:
        # Get null count using robust helper function
        null_series = df_renamed[col].isna().sum()
        null_count = get_scalar_null_count(null_series)
        null_pct = (null_count / len(df_renamed)) * 100
        if null_pct > 50:  # More than 50% null
            high_null_cols.append((col, null_count, null_pct))
    
    if high_null_cols:
        print(f"\n⚠️ Columns with >50% null values:")
        for col, null_count, null_pct in sorted(high_null_cols, key=lambda x: x[2], reverse=True):
            print(f"  {col}: {null_count:,} nulls ({null_pct:.1f}%)")
    
    # Check for critical missing columns
    schema_columns = set(processor.schema_data.get('enhanced_schema', {}).get('columns', {}).keys())
    available_columns = set(df_renamed.columns)
    missing_columns = schema_columns - available_columns
    
    if missing_columns:
        print(f"\n⚠️ Missing Schema Columns ({len(missing_columns)}):")
        for col in sorted(missing_columns):
            print(f"  - {col}")
    else:
        print(f"\n✅ All schema columns found in data!")
    
    cell_done("Excel data loaded, cleaned, and column mapping completed")
    
except Exception as e:
    print(f"❌ Error loading Excel data: {e}")
    import traceback
    traceback.print_exc()
    raise
```

```python
# Apply universal document processing
print("=" * 80)
print("\n⚙️ Step 3: Universal Document Processing")
print("=" * 50)

try:
    # Force reload the processor module to get the latest fixes
    import importlib
    import universal_document_processor
    importlib.reload(universal_document_processor)
    from universal_document_processor import UniversalDocumentProcessor
    
    # Recreate processor with updated module
    processor = UniversalDocumentProcessor()
    processor.schema_file = str(schema_path)
    processor.load_schema()
    print("✅ Reloaded universal document processor with latest fixes")
    
    # Use df_renamed from Step 2 (already mapped)
    print("📋 Using mapped dataframe from Step 2...")
    print(f"  Input dataframe: {df_renamed.shape}")
    
    # Show sample of columns
    print(f"\n📋 Available Columns ({len(df_renamed.columns)}):")
    for i, col in enumerate(df_renamed.columns[:10]):
        print(f"  {i+1:2d}. {col}")
    if len(df_renamed.columns) > 10:
        print(f"  ... and {len(df_renamed.columns) - 10} more columns")
    
    # Check for missing Document_Title column
    if 'Document_Title' not in df_renamed.columns:
        print(f"\n⚠️ Document_Title not found in data. Available similar columns:")
        similar_cols = [col for col in df_renamed.columns if 'document' in col.lower() or 'title' in col.lower() or 'description' in col.lower()]
        for col in similar_cols:
            print(f"  - {col}")
    
    # Process data with universal processor
    print("\n🔄 Applying universal processing...")
    df_processed = processor.process_data(df_renamed)
    
    print(f"\n📊 Processing Results:")
    print(f"  Input Shape: {df_renamed.shape}")
    print(f"  Output Shape: {df_processed.shape}")
    
    # Show sample of processed data
    print("\n📋 Sample Processed Data (first 5 rows):")
    display(df_processed.head())
    
    # Show processing summary
    print("\n📈 Processing Summary:")
    for col in df_processed.columns:
        null_count = df_processed[col].isna().sum()
        if null_count > 0:
            print(f"  {col}: {len(df_processed)} rows, {null_count} null values ⚠️")
        else:
            print(f"  {col}: {len(df_processed)} rows, {null_count} null values ✅")
    
    # Show validation results
    print("\n🔍 Validation Summary:")
    enhanced_schema = processor.schema_data.get('enhanced_schema', {})
    columns = enhanced_schema.get('columns', {})
    validation_results = {}
    
    for col_name, col_def in columns.items():
        if col_name in df_processed.columns:
            validation = col_def.get('validation', {})
            issues = []
            
            # Check pattern validation
            if 'pattern' in validation:
                pattern = validation['pattern']
                df_str = df_processed[col_name].astype(str)
                invalid_mask = ~df_str.str.match(pattern, na=False)
                invalid_count = invalid_mask.sum()
                if invalid_count > 0:
                    issues.append(f"Pattern: {invalid_count} invalid")
            
            # Check min length validation
            if 'min_length' in validation:
                min_len = validation['min_length']
                df_str = df_processed[col_name].astype(str)
                invalid_mask = df_str.str.len() < min_len
                invalid_count = invalid_mask.sum()
                if invalid_count > 0:
                    issues.append(f"Min length: {invalid_count} too short")
            
            # Check allowed values validation
            if 'allowed_values' in validation:
                allowed = validation['allowed_values']
                invalid_mask = ~df_processed[col_name].isin(allowed)
                invalid_count = invalid_mask.sum()
                if invalid_count > 0:
                    issues.append(f"Allowed values: {invalid_count} invalid")
            
            if issues:
                validation_results[col_name] = issues
                print(f"  {col_name}: {', '.join(issues)} ⚠️")
    
    if not validation_results:
        print("  All columns passed validation ✅")
    
except Exception as e:
    print(f"❌ Error during processing: {e}")
    import traceback
    traceback.print_exc()
```
